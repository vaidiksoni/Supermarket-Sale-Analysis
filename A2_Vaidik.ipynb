we compare the correlation between features and remove one of two features that have a correlation higher than 0.85

```python
columns = np.full((corr.shape[0],), True, dtype=bool)
for i in range(corr.shape[0]):
    for j in range(i+1, corr.shape[0]):
        if corr.iloc[i,j] >= 0.85:
            if columns[j]:
                columns[j] = False
selected_columns = df.columns[columns]
# df= df[selected_columns]
```

## Feature Scaling and applying PCA to further extract important columns

```python
from sklearn.preprocessing import scale
from sklearn.decomposition import PCA
```

```python
df.head()
```

```python
x = df.iloc[:,1:]
y = df.iloc[:,0]
```

```python
x = scale(x)
```

```python
#Plotting the Cumulative Summation of the Explained Variance


pca = PCA().fit(x)
plt.figure()
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Number of Components')
plt.ylabel('Variance (%)') #for each component
```

Now by looking at above curve we'll get 90%+ variance by using 12 components

```python
pca = PCA(n_components=12)
x = pca.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.25, random_state = 100)
```

# Model Selection
We'll first select Linear Regression

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(x_train, y_train)
```

```python
y_pred_lr = lr.predict(x_test)
```

```python
y_pred_lr
```

Model

```python
print('Intercept is ', lr.intercept_)
print('Coefficient is ',lr.coef_)
```



### Model Evaluation

```python
import statsmodels.api as sm
x_train_sm =x_train
x_train_sm = sm.add_constant(x_train_sm)
lm1 = sm.OLS(y_train, x_train_sm).fit()
```

```python
print(lm1.summary())
```

```python
# Calculatin error terms
import math
from sklearn.metrics import mean_squared_error, r2_score
mse = mean_squared_error(y_test, y_pred_lr)
r_sq = r2_score(y_test, y_pred_lr)
print('RMSE', math.sqrt(mse))
print('R Sq. value', r_sq)
```
